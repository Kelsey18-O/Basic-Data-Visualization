# Basic Data Visualization of Customer Metrics

## Introduction
This notebook explores customer data and analyzes relationships between different features and the 'Yearly Amount Spent'.
Using 'pandas', 'seaborn' and 'matplotlib', we will create various visualizations to uncover insights about customer behaviour.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset with column names
df = pd.read_csv('basic_data_visualiztion.csv')

# Check the data to confirm columns
print(df.head())
```

## Data Overview
The dataset contains the following columns:
1. 'Avg. Session Length'
2. 'Time on App'
3. 'Length of Membership'
4. 'Yearly Amount Spent'

The following table displays the first few rows of the dataset to provide an initial look.

```python
# Display first few rows of the data
print(df.head())
```

## Visualization 1: Bar Chart - Average Session vs Yearly Amount Spent 

This bar chart shows the relationship between the average session length and yearly amount spent by customers. The purpose is to
identify if there's any pattern or trend that could indicate a relationship between session length and spending.


```python
# Visualization 1: Bar Chart - Average Session Length vs Yearly Amount Spent
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Avg. Session Length', y='Yearly Amount Spent')
plt.title('Yearly Amount Spent by Average Session Length')
plt.xlabel('Average Session Length')
plt.ylabel('Yearly Amount Spent')
plt.show()
```

### From this bar chart, we can observe that a longer average session length is generally associated with higher yearly spending.

## Visualization 2: Line Chart - Time on App vs Yearly Amount Spent

The line chart below illustrates how the amount spent by customers varies with the time spent on the app. The visualization helps in understanding if there is a trend where increased app usage correlates with more spending.  

```python
# Visualization 2: Line Chart - Time on App vs Yearly Amount Spent
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Time on App', y='Yearly Amount Spent')
plt.title('Yearly Amount Spent by Time on App')
plt.xlabel('Time on App')
plt.ylabel('Yearly Amount Spent')
plt.show()
```

### Observation 
This line chart show a positive relationship between time on app and yearly spending indicating that customers who spend more time on the app tend to spend more anaully.


## Visualization 3: Scatter Plot - Length of Membership vs Yearly Amount Spent 

The scatter plot below examines the realtionship between the length of membership and the yearly amount spent. By examining this relationship, we can determine if a longer membership duration is associated with increased spending.  

```python
# Visualization 3: Scatter Plot - Length of Membership vs Yearly Amount Spent
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Length of Membership', y='Yearly Amount Spent')
plt.title('Yearly Amount Spent by Length of Membership')
plt.xlabel('Length of Membership')
plt.ylabel('Yearly Amount Spent')
plt.show()
```

## Observation
The scatter plot indicates a positive correlation, where customers with longer memberships generally spend more anually. This may suggest a relationship between customer loyalty and spending behaviour. 

## Conclusion
In this notebook, we explored various customer metrics and their relationship with yearly spending through different types of visualizations. Some key insights include:
- Customers with longer average session lengths and more time spent on app tend to spend more yearly. 
- Insights to inform customer about engagement strategies and loyaty programs to enhance customer lifetime value.
